##### 数据处理

```python
# 该函数作用
# 去除无效变量,对部分变量进行归一化
# 改变部分变量编码格式,尝试自定义函数
# 去除缺失行

def nomalizied(fn, col):
    labels = fn[col].unique().tolist()
#     fn[col] = [*map(lambda x: labels.index(x), fn[col])]
    fn[col] = fn[col].apply(lambda n: labels.index(n)) 

def info_processing(fn):
    data = pd.read_csv(fn)
    data.drop(['PassengerId', 'Name', 'Cabin', 'Ticket'], axis = 1, inplace = True)
    nomalizied(data, 'Sex')
    nomalizied(data, 'Embarked')
#     fn['Fare'].apply(int)
#     temp_col = fn['Fare']
#     fn['Fare'] = [int(x) for x in temp_col]\
    data = data.dropna()
#     data['Fare'] = data['Fare'].astype(int)
    data['Age'] = data['Age'].astype(int)
    
    return data
```

```python
train_pc = info_processing(train_add)
test_pc = info_processing(test_add)
```

##### 模型优化

######  考察深度

```python
x = train_pc.drop('Survived', axis = 1).values
y = train_pc['Survived'].values

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
# 设置树深度评估模型
from sklearn.tree import DecisionTreeClassifier
def cv_score(d):
    clt = DecisionTreeClassifier(max_depth = d)
    clt.fit(x_train, y_train)
    
    return(clt.score(x_train, y_train), clt.score(x_test, y_test))
```

```python
depths = np.arange(1,15)
s = [cv_score(i) for i in depths]
tr_score = [i[0] for i in s]
te_score = [i[1] for i in s]

tr_best_index = np.argmax(tr_score)
te_best_index = np.argmax(te_score)

print('测试集中最佳深度', te_best_index + 1, '\n', '测试集最佳分数', te_score[tr_best_index])
```

```python
%matplotlib inline

from matplotlib import pyplot as plt
plt.figure(figsize = (14, 8), dpi = 120)
plt.grid() #网格线
plt.xlabel('deepth of the tree')
plt.ylabel('score')
plt.plot(depths, tr_score, label = 'train_score')
plt.plot(depths, te_score, label = 'test_score')
plt.legend() #分类
```

###### 考察基尼不纯度

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)

# 针对基尼不纯的最优值测试
def minsplit_score(val):
    clf = DecisionTreeClassifier(criterion = 'gini', min_impurity_decrease = val)
    clf.fit(x_train, y_train)
    return(clf.score(x_train, y_train), clf.score(x_test, y_test))
```

```python
vals = np.linspace(0, 0.2, 20)
s = [minsplit_score(i) for i in vals]
tr_score = [i[0] for i in s]
te_score = [i[1] for i in s]

tr_best_index = np.argmax(tr_score)
te_best_index = np.argmax(te_score)

print('best gini value of test', vals[te_best_index], 'best score of test', te_score[te_best_index])
```

```python
plt.figure(figsize = (10, 4), dpi = 120)
plt.grid()
plt.xlabel('gini values')
plt.ylabel('score')
plt.plot(vals, tr_score, label = 'train data')
plt.plot(vals, te_score, label = 'test data') #前两个参数为数组名
plt.legend()
```
